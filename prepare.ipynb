# Superstore

Using your store items data:

1) Convert date column to datetime format.

```python
store = pd.read_csv('ts_superstore.csv', index_col=0)
```

```python
store.head()
```

```python
store.sale_date = store.sale_date.str.replace(' 00:00:00 GMT', '')
```

```python
store.sale_date = pd.to_datetime(store.sale_date, format='%a, %d %b %Y')
```

```python
store.head()
```

2) Plot the distribution of sale_amount and item_price.

```python
plt.figure(figsize=(15,6))

plt.subplot(121)
sns.histplot(store.sale_amount)

plt.subplot(122)
sns.histplot(store.item_price)

plt.show()
```

3) Set the index to be the datetime variable.

```python
store = store.set_index('sale_date')
```

```python
store.head()
```

4) Add a 'month' and 'day of week' column to your dataframe.

```python
store['month'] = store.index.month_name()
store['day_of_week'] = store.index.day_name()

store.head()
```

5) Add a column to your dataframe, sales_total, which is a derived from sale_amount (total items) and item_price.

```python
store['sales_total'] = store.sale_amount * store.item_price
```

```python
store[['sales_total', 'sale_amount', 'item_price']]
```
